```python
Gene2exonCount = pd.DataFrame()
Gene2exonCount['GeneName'] = None
Gene2exonCount['ExonCount'] = None
```

```python
Gene2exonCount['GeneName'] = ['g' + str(i) for i in range(1, 64456)]
Gene2exonCount
```

```python
filtered_df = braker[(braker[2] == 'exon')]
filtered_df
```

```python
NR_braker = filtered_df[~filtered_df.duplicated(subset=[3, 4], keep='first')]

# 결과 확인
NR_braker
```

```python
for i in range(1, 64456):
    value_to_search = f'"g{i}"'
    filtered_rows = NR_braker[NR_braker[8].str.contains(value_to_search)]
    Gene2exonCount = Gene2exonCount.append({'ExonCount': len(filtered_rows)}, ignore_index=True)
```

```python
k = Gene2exonCount["ExonCount"].dropna().reset_index(drop=True)
```

```python
r = Gene2exonCount.iloc[:64455,:]
```

```python
r["Exon_Count"] = k
```

```python
r = r.drop('ExonCount',axis=1)
```

```python
r.to_csv("/Users/csb/Desktop/ExonCount.csv",index=False)
```
